```python
df_sales=pd.read_csv('sales_data_final (1).csv')
df_sales.head()
```

```python
def profit_flag(a):
    if(a>0):
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(45)
# print(a)
```

```python
df_sales['flag_profit']=df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

```python
df_sales['len_cust_name']=df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
def quantity_flag(a):
    if a<10:
        return 'low'
    elif a>10 and a<20:
        return 'medium'
    else:
        return 'high'
```

```python
df_sales['flag_quantity']=df_sales['quantity'].apply(quantity_flag)
```
