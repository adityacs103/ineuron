```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
population_size=1000000
population=np.random.rand(1000000)
```

```python
number_of_sample=10000
sample_mean=np.random.rand(number_of_sample)
sample_size=50
```

```python
c=np.random.rand(number_of_sample)
for i in range(0,number_of_sample):
    c=np.random.randint(1,population_size,sample_size)
    sample_mean[i]=population[c].mean()
```

```python
plt.subplot(1,2,1)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.distplot(sample_mean,bins=int(180/5),hist=True,kde=False)
plt.title("histogram of sample mean",fontsize=20)
plt.xlabel("sample mean",fontsize=20)
plt.ylabel("count",fontsize=20)
plt.subplot(1,2,2)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.distplot(sample_mean,hist=True,kde=True)
plt.title("Density of sample mean",fontsize=20)
plt.xlabel("sample mean",fontsize=20)
plt.ylabel("Density",fontsize=20)
plt.subplots_adjust(bottom=0.1,right=2,top=0.9)
```
